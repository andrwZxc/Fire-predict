K-fold + обучение моделей

Так как у нас 3 столбца под метки (трехмерный массив), то необходимо использовать многозадачную классификацию MultiOutputClassifier

данные делятся на 5 фолдов

4 фолда → тренировочные данные (X_fold_train, y_fold_train).

1 фолд → валидационные данные (X_val, y_val).

В итоге модель обучается 5 раз на разных подмножествах данных, и каждый раз проверяется на отдельном фолде. Каждый из 5 фолдов побывает в роли валидационного.

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans

# Кодируем метки
label_encoders = {}
y_train_encoded = y_train.copy()
for col in y_train.columns:
    le = LabelEncoder()
    y_train_encoded[col] = le.fit_transform(y_train[col].astype(str))  # Преобразуем все в строки перед кодированием
    label_encoders[col] = le

models = {
    #"Logistic Regression": MultiOutputClassifier(LogisticRegression()),
    #"Random Forest": RandomForestClassifier(random_state=42),
    #"SVM": MultiOutputClassifier(SVC(probability=True)),
    #"KNN": MultiOutputClassifier(KNeighborsClassifier()),
    "k-means": KMeans()
}

kf = KFold(n_splits=5, shuffle=False)

for name, model in models.items():
    print(f"\n=== Модель: {name} ===")

    fold_scores = []
    for train_idx, val_idx in kf.split(X_train_scaled):
        X_fold_train, X_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
        y_fold_train, y_val = y_train_encoded.iloc[train_idx], y_train_encoded.iloc[val_idx]  # Используем закодированные метки

        model.fit(X_fold_train, y_fold_train)
        y_pred = model.predict(X_val)

        # Средняя accuracy по всем меткам
        score = np.mean([accuracy_score(y_val.iloc[:, i], y_pred[:, i]) for i in range(y_val.shape[1])])
        fold_scores.append(score)
        print(f"Fold accuracy: {score:.4f}")

    print(f"Средняя точность (CV): {np.mean(fold_scores):.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical


train_data = pd.read_csv("TRAIN_PASCAL.csv").fillna(0).drop(['Date', 'Sensor_ID', 'Group'], axis=1)
test_data = pd.read_csv("TEST_PASCAL.csv").fillna(0).drop(['Date', 'Sensor_ID', 'Group'], axis=1)

X_train = train_data.drop(columns=["fire", "scenario_label", "progress_label"])
y_train = train_data["fire"]  # Бинарная классификация - только факт возгорания

# Масштабирование
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(test_data.drop(columns=["fire", "scenario_label", "progress_label"]))

# Преобразование в 3D-формат для LSTM [samples, timesteps, features]
# Предположим, у нас есть временные последовательности длиной 10 измерений
timesteps = 10
features = X_train_scaled.shape[1]

# Создание последовательностей
def create_sequences(X, y, timesteps):
    Xs, ys = [], []
    for i in range(len(X) - timesteps):
        Xs.append(X[i:(i + timesteps)])
        ys.append(y[i + timesteps])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_sequences(X_train_scaled, y_train.values, timesteps)

# Разделение на train/val
X_train_seq, X_val_seq, y_train_seq, y_val_seq = train_test_split(
    X_seq, y_seq, test_size=0.2, random_state=42
)

# Построение модели LSTM
model = Sequential([
    LSTM(64, input_shape=(timesteps, features), return_sequences=True),
    Dropout(0.2),
    LSTM(32),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Обучение
history = model.fit(
    X_train_seq, y_train_seq,
    validation_data=(X_val_seq, y_val_seq), #
    epochs=20,
    batch_size=32
)

# Оценка
test_loss, test_acc = model.evaluate(X_val_seq, y_val_seq)
print(f"\nTest Accuracy: {test_acc:.4f}")

# Прогнозирование
predictions = model.predict(X_val_seq)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score
from sklearn.pipeline import Pipeline


y_train = np.ravel(y_train)  # Преобразуем в 1D массив
y_test = np.ravel(y_test)

# 4. TimeSeriesSplit только для тренировочных данных
tscv = TimeSeriesSplit(n_splits=5)

# 5. Pipeline с нормализацией и моделью
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(random_state=42))
])

# 6. Кросс-валидация на тренировочных данных
f1_scores = []
roc_auc_scores = []

for fold_num, (train_index, val_index) in enumerate(tscv.split(X_train), 1):
    try:
        # Разделение на тренировочную и валидационную части
        X_tr, X_val = X_train.iloc[train_index], X_train.iloc[val_index]
        y_tr, y_val = y_train[train_index], y_train[val_index]

        # Обучение и предсказание
        pipeline.fit(X_tr, y_tr)
        y_proba = pipeline.predict_proba(X_val)[:, 1]
        y_pred = pipeline.predict(X_val)

        # Расчет метрик
        f1 = f1_score(y_val, y_pred)
        roc_auc = roc_auc_score(y_val, y_proba)

        f1_scores.append(f1)
        roc_auc_scores.append(roc_auc)

        print(f"Фолд {fold_num}: F1={f1:.4f}, ROC-AUC={roc_auc:.4f}")
        print(f"Размеры: train={len(X_tr)}, val={len(X_val)}")

    except Exception as e:
        print(f"Ошибка в фолде {fold_num}: {str(e)}")
        continue

# 7. Финальная оценка на тестовом наборе
if f1_scores:
    print("\nСредние метрики на валидации:")
    print(f"F1-score: {np.mean(f1_scores):.4f} (±{np.std(f1_scores):.4f})")
    print(f"ROC-AUC: {np.mean(roc_auc_scores):.4f} (±{np.std(roc_auc_scores):.4f})")

    # Обучение на всех тренировочных данных
    pipeline.fit(X_train, y_train)

    # Предсказание на тестовых данных
    test_proba = pipeline.predict_proba(X_test)[:, 1]
    test_pred = pipeline.predict(X_test)

    # Метрики на тесте
    test_f1 = f1_score(y_test, test_pred)
    test_roc_auc = roc_auc_score(y_test, test_proba)

    print("\nФинальные метрики на тесте:")
    print(f"F1-score: {test_f1:.4f}")
    print(f"ROC-AUC: {test_roc_auc:.4f}")
else:
    print("Не удалось вычислить метрики ни для одного фолда")
```

```python
import pickle
from datetime import datetime

# 1. Сохранение модели
def save_model(model, X_train, metrics=None):
    # Создаем словарь с моделью и метаданными
    model_data = {
        'model': model,
        'metadata': {
            'training_date': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            'features_used': list(X_train.columns),
            'metrics': {
                'f1_score': np.mean(f1_scores) if f1_scores else None,
                'roc_auc': np.mean(roc_auc_scores) if roc_auc_scores else None,
                'test_f1': test_f1 if 'test_f1' in locals() else None,
                'test_roc_auc': test_roc_auc if 'test_roc_auc' in locals() else None
            }
        }
    }

    # Имя файла с timestamp
    filename = f"model_diplom.pkl"

    # Сохраняем в файл
    with open(filename, 'wb') as f:
        pickle.dump(model_data, f)

    print(f"Модель сохранена как {filename}")
    return filename

# 2. Сохраняем нашу модель (после обучения на всех данных)
if f1_scores:
    model_filename = save_model(pipeline, X_train)
else:
    print("Модель не обучена, сохранение невозможно")
```
