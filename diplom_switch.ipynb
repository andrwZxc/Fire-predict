```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, f1_score, recall_score, precision_score, accuracy_score
from sklearn.pipeline import Pipeline

def evaluate_model(pipeline, model_name, X_train, y_train, X_test, y_test):
    print(f"\n=== Оценка модели {model_name} ===")

    # Проверка и сохранение имен признаков
    feature_names = X_train.columns.tolist()
    print(f"Используемые признаки ({len(feature_names)}):", feature_names)

    # Инициализация кросс-валидации
    tscv = TimeSeriesSplit(n_splits=3)  # Уменьшил число фолдов для скорости

    f1_scores = []
    roc_auc_scores = []
    recall_scores = []
    precision_scores = []
    accuracy_scores = []

    for fold_num, (train_index, val_index) in enumerate(tscv.split(X_train), 1):
        try:
            # Получаем данные с сохранением структуры DataFrame
            X_tr = X_train.iloc[train_index].reset_index(drop=True)
            X_val = X_train.iloc[val_index].reset_index(drop=True)
            y_tr = y_train.iloc[train_index].reset_index(drop=True)
            y_val = y_train.iloc[val_index].reset_index(drop=True)

            # Обучение
            pipeline.fit(X_tr, y_tr)

            # Предсказание (используем только нужные признаки)
            try:
                y_proba = pipeline.predict_proba(X_val[feature_names])[:, 1]
                y_pred = pipeline.predict(X_val[feature_names])
            except KeyError as e:
                print(f"Ошибка в признаках: {str(e)}")
                continue

            # Расчет метрик
            f1 = f1_score(y_val, y_pred)
            roc_auc = roc_auc_score(y_val, y_proba)
            recall = recall_score(y_val, y_pred, zero_division=0)
            precision = precision_score(y_val, y_pred, zero_division=0)
            accuracy = accuracy_score(y_val, y_pred)

            # Сохранение результатов
            f1_scores.append(f1)
            roc_auc_scores.append(roc_auc)
            recall_scores.append(recall)
            precision_scores.append(precision)
            accuracy_scores.append(accuracy)

            print(f"\nФолд {fold_num}:")
            print(f"F1={f1:.4f}, ROC-AUC={roc_auc:.4f}")
            print(f"Recall={recall:.4f}, Precision={precision:.4f}")
            print(f"Accuracy={accuracy:.4f}")
            print(f"Размеры: train={len(X_tr)}, val={len(X_val)}")

        except Exception as e:
            print(f"\nОшибка в фолде {fold_num}: {str(e)}")
            continue

    if f1_scores:
        print("\n=== Итоговые метрики на валидации ===")
        print(f"F1-score: {np.mean(f1_scores):.4f} (±{np.std(f1_scores):.4f})")
        print(f"ROC-AUC: {np.mean(roc_auc_scores):.4f} (±{np.std(roc_auc_scores):.4f})")
        print(f"Recall: {np.mean(recall_scores):.4f} (±{np.std(recall_scores):.4f})")
        print(f"Precision: {np.mean(precision_scores):.4f} (±{np.std(precision_scores):.4f})")
        print(f"Accuracy: {np.mean(accuracy_scores):.4f} (±{np.std(accuracy_scores):.4f})")

        # Финальное обучение на всех данных
        pipeline.fit(X_train[feature_names], y_train)

        # Предсказание на тесте
        test_proba = pipeline.predict_proba(X_test[feature_names])[:, 1]
        test_pred = pipeline.predict(X_test[feature_names])

        print("\n=== Метрики на тестовом наборе ===")
        print(f"F1-score: {f1_score(y_test, test_pred):.4f}")
        print(f"ROC-AUC: {roc_auc_score(y_test, test_proba):.4f}")
        print(f"Recall: {recall_score(y_test, test_pred):.4f}")
        print(f"Precision: {precision_score(y_test, test_pred):.4f}")
        print(f"Accuracy: {accuracy_score(y_test, test_pred):.4f}")
    else:
        print("Не удалось вычислить метрики ни для одного фолда")

# Создание пайплайнов
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        random_state=42,
        class_weight='balanced'
    ))
])

lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression(
        max_iter=1000,
        random_state=42,
        class_weight='balanced',
        penalty='l2',
        C=1.0
    ))
])

# Проверка данных перед обучением
print("\nПроверка данных:")
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
print("Примеры признаков:", X_train.columns.tolist()[:5])

# Запуск оценки моделей
evaluate_model(rf_pipeline, "RandomForest", X_train, y_train, X_test, y_test)
evaluate_model(lr_pipeline, "LogisticRegression", X_train, y_train, X_test, y_test)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, recall_score
from sklearn.pipeline import Pipeline

# Инициализация кросс-валидации
tscv = TimeSeriesSplit(n_splits=5)

# Создание pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Списки для хранения метрик
f1_scores = []
roc_auc_scores = []
recall_scores = []

# Кросс-валидация
for fold_num, (train_index, val_index) in enumerate(tscv.split(X_train), 1):
    try:
        # Разделение данных
        X_tr, X_val = X_train.iloc[train_index], X_train.iloc[val_index]
        y_tr, y_val = y_train[train_index], y_train[val_index]

        # Обучение и предсказание
        pipeline.fit(X_tr, y_tr)
        y_proba = pipeline.predict_proba(X_val)[:, 1]
        y_pred = pipeline.predict(X_val)

        # Расчет метрик
        f1 = f1_score(y_val, y_pred)
        roc_auc = roc_auc_score(y_val, y_proba)
        recall = recall_score(y_val, y_pred, average='weighted')  # Взвешенный recall

        # Сохранение метрик
        f1_scores.append(f1)
        roc_auc_scores.append(roc_auc)
        recall_scores.append(recall)

        print(f"Фолд {fold_num}: F1={f1:.4f}, ROC-AUC={roc_auc:.4f}, Recall(w)={recall:.4f}")
        print(f"Размеры: train={len(X_tr)}, val={len(X_val)}")

    except Exception as e:
        print(f"Ошибка в фолде {fold_num}: {str(e)}")
        continue

# Вывод средних метрик валидации
if f1_scores:
    print("\nСредние метрики на валидации:")
    print(f"F1-score: {np.mean(f1_scores):.4f} (±{np.std(f1_scores):.4f})")
    print(f"ROC-AUC: {np.mean(roc_auc_scores):.4f} (±{np.std(roc_auc_scores):.4f})")
    print(f"Recall(w): {np.mean(recall_scores):.4f} (±{np.std(recall_scores):.4f})")

    # Финальное обучение и оценка на тесте
    pipeline.fit(X_train, y_train)
    test_proba = pipeline.predict_proba(X_test)[:, 1]
    test_pred = pipeline.predict(X_test)

    print("\nФинальные метрики на тесте:")
    print(f"F1-score: {f1_score(y_test, test_pred):.4f}")
    print(f"ROC-AUC: {roc_auc_score(y_test, test_proba):.4f}")
    print(f"Recall(w): {recall_score(y_test, test_pred, average='weighted'):.4f}")
else:
    print("Не удалось вычислить метрики ни для одного фолда")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, roc_auc_score, f1_score
from sklearn.pipeline import Pipeline
from imblearn.ensemble import BalancedRandomForestClassifier
from imblearn.pipeline import make_pipeline


# Настройка кросс-валидации для временных рядов
tscv = TimeSeriesSplit(n_splits=5)

# Создаем кастомные метрики для кросс-валидации
scoring = {
    'f1': make_scorer(f1_score),
    'roc_auc': make_scorer(roc_auc_score, needs_proba=True),
    'f1_weighted': make_scorer(f1_score, average='weighted')
}

# Вариант 1: Сбалансированный Random Forest (рекомендуется при дисбалансе)
pipeline = make_pipeline(
    StandardScaler(),
    BalancedRandomForestClassifier(
        random_state=42,
        sampling_strategy='auto',  # автоматический баланс классов
        class_weight='balanced_subsample'  # баланс в каждом дереве
    )
)


# Кросс-валидация с использованием встроенных методов
cv_results = cross_validate(
    pipeline,
    X_train,
    y_train,
    cv=tscv,
    scoring=scoring,
    return_train_score=True,
    n_jobs=-1  # параллелизация
)

# Вывод результатов кросс-валидации
print("Результаты кросс-валидации:")
print(f"Средний F1-score: {np.mean(cv_results['test_f1']):.4f} (±{np.std(cv_results['test_f1']):.4f})")
print(f"Средний ROC-AUC: {np.mean(cv_results['test_roc_auc']):.4f} (±{np.std(cv_results['test_roc_auc']):.4f})")
print(f"Средний Weighted F1: {np.mean(cv_results['test_f1_weighted']):.4f}")

# Обучение на всех тренировочных данных
pipeline.fit(X_train, y_train)

# Оценка на тестовых данных
test_proba = pipeline.predict_proba(X_test)[:, 1]
test_pred = pipeline.predict(X_test)

# Расчет метрик с учетом дисбаланса
test_f1 = f1_score(y_test, test_pred)
test_roc_auc = roc_auc_score(y_test, test_proba)
test_f1_weighted = f1_score(y_test, test_pred, average='weighted')

print("\nФинальные метрики на тесте:")
print(f"F1-score: {test_f1:.4f}")
print(f"ROC-AUC: {test_roc_auc:.4f}")
print(f"Weighted F1-score: {test_f1_weighted:.4f}")

# Дополнительно: можно вывести важность признаков
if hasattr(pipeline.named_steps['balancedrandomforestclassifier'], 'feature_importances_'):
    importances = pipeline.named_steps['balancedrandomforestclassifier'].feature_importances_
    features = X_train.columns
    print("\nВажность признаков:")
    for feature, importance in sorted(zip(features, importances), key=lambda x: x[1], reverse=True):
        print(f"{feature}: {importance:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt


y_train = np.ravel(y_train)  # Преобразуем в 1D массив
y_test = np.ravel(y_test)

# 4. TimeSeriesSplit только для тренировочных данных
tscv = TimeSeriesSplit(n_splits=5)

# 5. Pipeline с нормализацией и моделью
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier(random_state=42))
])

# 6. Кросс-валидация на тренировочных данных
f1_scores = []
roc_auc_scores = []


# 6. Кросс-валидация на тренировочных данных с визуализацией
plt.figure(figsize=(10, 8))
f1_scores = []
roc_auc_scores = []

# Цвета для каждого фолда
colors = ['blue', 'green', 'red', 'cyan', 'magenta']

for fold_num, (train_index, val_index) in enumerate(tscv.split(X_train), 1):
    try:
        # Разделение на тренировочную и валидационную части
        X_tr, X_val = X_train.iloc[train_index], X_train.iloc[val_index]
        y_tr, y_val = y_train[train_index], y_train[val_index]

        # Обучение и предсказание
        pipeline.fit(X_tr, y_tr)
        y_proba = pipeline.predict_proba(X_val)[:, 1]
        y_pred = pipeline.predict(X_val)

        # Расчет метрик
        f1 = f1_score(y_val, y_pred)
        roc_auc = roc_auc_score(y_val, y_proba)

        f1_scores.append(f1)
        roc_auc_scores.append(roc_auc)

        # ROC кривая для текущего фолда
        fpr, tpr, _ = roc_curve(y_val, y_proba)
        roc_auc = auc(fpr, tpr)

        plt.plot(fpr, tpr, color=colors[fold_num-1], lw=2,
                 label='Фолд %d (AUC = %0.2f)' % (fold_num, roc_auc))

        print(f"Фолд {fold_num}: F1={f1:.4f}, ROC-AUC={roc_auc:.4f}")
        print(f"Размеры: train={len(X_tr)}, val={len(X_val)}")

    except Exception as e:
        print(f"Ошибка в фолде {fold_num}: {str(e)}")
        continue

# Добавляем случайный классификатор (baseline)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Случайный классификатор')

# Настройки графика
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC кривые по фолдам кросс-валидации')
plt.legend(loc="lower right")
plt.grid()
plt.show()

# 7. Финальная оценка на тестовом наборе с ROC кривой
if f1_scores:
    print("\nСредние метрики на валидации:")
    print(f"F1-score: {np.mean(f1_scores):.4f} (±{np.std(f1_scores):.4f})")
    print(f"ROC-AUC: {np.mean(roc_auc_scores):.4f} (±{np.std(roc_auc_scores):.4f})")

    # Обучение на всех тренировочных данных
    pipeline.fit(X_train, y_train)

    # Предсказание на тестовых данных
    test_proba = pipeline.predict_proba(X_test)[:, 1]
    test_pred = pipeline.predict(X_test)

    # Метрики на тесте
    test_f1 = f1_score(y_test, test_pred)
    test_roc_auc = roc_auc_score(y_test, test_proba)

    # ROC кривая для тестовых данных
    fpr_test, tpr_test, _ = roc_curve(y_test, test_proba)
    roc_auc_test = auc(fpr_test, tpr_test)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr_test, tpr_test, color='darkorange', lw=2,
             label='Тест (AUC = %0.2f)' % roc_auc_test)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC кривая на тестовых данных')
    plt.legend(loc="lower right")
    plt.grid()
    plt.show()

    print("\nФинальные метрики на тесте:")
    print(f"F1-score: {test_f1:.4f}")
    print(f"ROC-AUC: {test_roc_auc:.4f}")
else:
    print("Не удалось вычислить метрики ни для одного фолда")
```
